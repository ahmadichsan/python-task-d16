**Common Pitfall when Performing EDA**
1. Statistical Fallacies

   a. Error in interpreting data resulting in inappropriate insights
   
   b. Sampling Bias => Mengambil kesimpulan dari state of data yang sebenarnya tidak representatif terhadap population. Contoh, sampling suka anjing/kucing tapi samplingnya dari orang2 yg suka anjing
   
   c. Cherry Picking => Mengambil data yang cocok dengan claim kita dan menyingkirkan data yang tidak cocok dengan claim kita
   
   d. False Causality => Asumsikan bahwa korelitas dari dua fenomena memiliki elemen kausalitas. Intinya, cocokologi wkwkwk
   
   e. Gambler's Falacy => Percaya secara salah, sesuatu yg sudah berlangsung terus menerus, maka kedepannya hal tsb akan terjadi lebih sedikit (dan kebalikannya). Contoh, gambler melihat bola sudah masuk ke hitam 26x. Karena udah sering hitam, maka bakal lebih sering yang putih nih. Nah ini salah. Kembali ke probabilitasnya
   
   f. Simpson's Paradox => Bisa mendapat insight yang berbeda saat melihat secara subset datanya aja dibanding saat melihat data secara keseluruhan
   
   g. Survivorship Bias => mengambil kesimpulan dari data yang tidak lengkap. Karena data yang ada ("survived") sudah melalui beberapa kriteria.
